```python
import numpy as sp
traffic = sp.genfromtxt("/content/drive/MyDrive/PP/web_traffic.tsv",delimiter='\t')
print(traffic[:10])
print(traffic.shape)
```

```python
x = traffic[:,0]
y = traffic[:,1]
```

```python
x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("Web traffic last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
```

```python
plt.xticks([w*7*24 for w in range(10)],['week %i' %w for w in range(10)])
plt.autoscale(tight=True)
plt.grid()
```
